```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.insert(0, '../carrot/')
from constants import *
from utils import *

methods = ['rorf',
           'mf',
           'carrot-knn',
           'carrot-knn-sbert',
           'carrot-roberta']

labels = {'mf':'RouteLLM (MF)',
          'carrot-knn':'CARROT (KNN)',
          'carrot-knn-sbert':'CARROT (KNN+SBERT)',
          'rorf':'Not-Diamond RoRF',
          'carrot-roberta':'CARROT (Roberta)'}

colors = {'mf':'blue',
          'carrot-knn':'red',
          'carrot-knn-sbert':'brown',
          'rorf':'green',
          'carrot-roberta':'orange'}
```

```python
dataset = 'routerbench' #'open-llm-lb-v2' #sprout'

data = np.load(f"../data/{dataset}/{dataset}_data_train_test_preds.npy", allow_pickle=True).item()
if 'Y_hat_carrot-knn-sbert ' in data.keys(): data['Y_hat_carrot-knn-sbert'] = data['Y_hat_carrot-knn-sbert ']
if 'OT_hat_carrot-knn-sbert ' in data.keys(): data['OT_hat_carrot-knn-sbert'] = data['OT_hat_carrot-knn-sbert ']

SMALL_MODEL, LARGE_MODEL = LARGE_SMALL_MODELS[dataset]['SMALL_MODEL'], LARGE_SMALL_MODELS[dataset]['LARGE_MODEL']
large_model_ind = int(np.argmax(np.array(data['models'])==LARGE_MODEL))
small_model_ind = int(np.argmax(np.array(data['models'])==SMALL_MODEL))
```

```python
data.keys()
```

```python
Y_test = data['Y_test']

if dataset == 'open-llm-lb-v2':
    mm = [m.replace("open-llm-leaderboard/","").replace("__","/").replace("-details","") for m in OPEN_MODELS]
    C_test = data['IT_test'][:,[np.argmax(np.array(data['models'])==m) for m in mm if np.sum(np.array(data['models'])==m)>0]]
    C_test = C_test*np.array([OPEN_COSTS[m]/1e6 for m in data['models']])[None,:]
    C_hat = C_test
elif dataset == 'sprout':
    input_costs = np.array([SPROUT_COSTS[m][0]/1e6 for m in data['models']])[None,:]
    output_costs = np.array([SPROUT_COSTS[m][1]/1e6 for m in data['models']])[None,:]
    C_test = input_costs*data['IT_test']+output_costs*data['OT_test']
elif dataset == 'routerbench':
    C_test = data['C_test']
else:
    raise(NotImplementedError)
```

```python
mult=100

routers = {}
for method in methods:

    routers[method] = {}
    if 'mf' in method or 'rorf' in method:
        routers[method]['cost'], routers[method]['perf'] = route_pairwise(data['Y_hat_'+method].squeeze(), C_test, Y_test, large_model_ind, small_model_ind)
    #elif 'lamb-' in model:
    #    routers[method]['cost'], routers[method]['perf'] = route_lamb(data['Y_hat_'+method], C_test, mult*C_hat, Y_test)
    else:  
        if dataset == 'open-llm-lb-v2':
            C_hat = C_test
        elif dataset == 'sprout':
            C_hat = input_costs*data['IT_test']+output_costs*data[f'OT_hat_{method}']
        elif dataset == 'routerbench':
            C_hat = data[f'C_hat_{method}'] 
        else:
            raise(NotImplementedError)
    
        routers[method]['cost'], routers[method]['perf'] = route(data['Y_hat_'+method], C_test, mult*C_hat, Y_test)
```

```python
models = data['models']
fig, ax = plt.subplots(1, 1, figsize = (4.3, 4.3))

markers = ['o', 's', 'D', '^', 'v', 'p', '*', 'x', '+', 'h', 'H', 'd', '>', 'P', '<', '|', '_', '.', ',', '1']

for method in methods:
    if 'lamb-' in method:
        if method.split('lamb-')[1]=='0':
            label = 'ERM router'
        else:
            label = ''
        ax.scatter(routers[method]['cost'], routers[method]['perf'], marker = markers[13], c='black', label = label)
    else:
        ax.errorbar(routers[method]['cost'], routers[method]['perf'], c=colors[method],
        linestyle='--', linewidth=1, alpha = 1, label = labels[method])

for i, txt in enumerate(models):

    x, y = C_test[:,i].mean(0), Y_test[:,i].mean(0)
    ax.scatter([x], [y], marker = markers[i])
    if dataset == 'open-llm-lb-v2':
        if i in set([3,4,6]):
            ax.annotate(models[i].split('/')[1], (x, y), size=7, horizontalalignment='right')
        if i in set([5,10,13,14,15]):
            ax.annotate(models[i].split('/')[1], (x, y), size=7)

ax.set_title(dataset)
ax.set_xlabel('Cost Per Query, \$')
ax.set_ylabel('Accuracy')
ax.legend()
ax.grid(True)
plt.show()
```
